### Общая информация

### Атрибуты
1. **Timestamp** (<font color="#240672">int</font>) - время начала временного окна (каждое окно по $60c$) в формате Unix
2. **Open** (<font color="#A68C00">float</font>) - цена на момент открытия временного окна
3. **High** (<font color="#A68C00">float</font>) - $max_{цена}$ во временном окне
4. **Low** (<font color="#A68C00">float</font>) - $min_{цена}$ во временном окне
5. **Close** (<font color="#A68C00">float</font>) - цена на момент закрытия временного окна
6. **Volume_(BTC)** (<font color="#A68C00">float</font>) - объём транзакции в $BTC$
7. **Volume_(Currency)** (<font color="#A68C00">float</font>) - объём транзакции в валюте
8. **Weighted_Price** (<font color="#A68C00">float</font>) - курс $BTC$ для данной валюты

## Опишу свои действия
- среднее значение атрибута $\overline{x}$ буду считать по формуле: $\overline{x}=\frac{1}{N}\sum\limits_{i=1}^{N}x_i$
    - `df['attribute'].mean()`, где `attribute` - название атрибута
- среднеквадратическое отклонение $s$ буду считать по формуле: $s=\sqrt{\frac{1}{N-1}\sum\limits_{i=1}^{N}(x_i-\overline{x})^2}$
    - `df['attribute'].std(ddof=1)`, где `attribute` - название атрибута
- `df.hist(column='attribute')`, где `attribute` - название атрибута
- выбросы
- количество пропущенных значений буду вычислять подсчётом количества ячеек, где в качестве значения указано `NaN`
    - `df['attribute'].isna().sum()`, где `attribute` - название атрибута
- вариантов обработки значений у меня два (зависит от типа анализа): отбросить или оставить
